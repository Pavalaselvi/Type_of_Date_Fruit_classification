# **SPLITTING DATA**

```python
# Assigning all rows of dataframe to the variables X & y
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Splitting dataframe into train & test data 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)

# Shape of the train & test data
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **DECISION TREE CLASSIFIER**

```python
# Building the model
clf_model = DecisionTreeClassifier(criterion="gini", random_state=42,max_depth=10, min_samples_leaf=10)   
clf_model.fit(X_train,y_train)
```

```python
# Predicting class
y_pred3 = clf_model.predict(X_test)
print(y_pred3)
```

```python
# Checking the accuracy 
print('Training score',clf_model.score(X_train,y_train))
print('Testing score',clf_model.score(X_test,y_test))
```

```python
# Confusion matrix & Classification report
cm3 = confusion_matrix(y_test, y_pred3) 
print ("Confusion Matrix : \n", cm3)
cr3 = classification_report(y_test, y_pred3)
print("\n Classification Report: \n", cr3)
```

# **RANDOM FOREST CLASSIFIER**

```python
# Building the model
clf = RandomForestClassifier(n_estimators = 100) 
clf.fit(X_train, y_train)
```

```python
# Predicting class
y_pred = clf.predict(X_test)
print(y_pred)
```

```python
# Checking the accuracy 
print('Training score',clf.score(X_train,y_train))
print('Testing score',clf.score(X_test,y_test))
```

```python
# Confusion matrix & Classification report
cm = confusion_matrix(y_test, y_pred) 
print ("Confusion Matrix : \n", cm)
cr = classification_report(y_test, y_pred)
print("\n Classification Report: \n", cr)
```

# **K NEAREST NEIGHBOUR CLASSIFICATION**

```python
# Building the model
m2 = KNeighborsClassifier(n_neighbors=25)
m2.fit(X_train,y_train)
```

```python
# Predicting class
y_pred2 = m2.predict(X_test)
print(y_pred2)
```

```python
# Checking the accuracy 
print('Training score',m2.score(X_train,y_train))
print('Testing score',m2.score(X_test,y_test))
```

```python
# Confusion matrix & Classification report
cm2 = confusion_matrix(y_test, y_pred2) 
print ("Confusion Matrix : \n", cm2)
cr2 = classification_report(y_test, y_pred2)
print("\n Classification Report: \n", cr2)
```
